```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('../instructor/Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values

```

```python
dataset.head()
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:] # removing col 0 for escaping dummy variable trap


```

```python
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# initialising the nn
```

```python
classifier = Sequential()
```

```python
# adding the first input layer and hidden layer/
classifier.add(Dense(output_dim=6, init='uniform', activation='relu', input_shape=(11,) ))
```

```python
classifier.add(Dense(output_dim=6,init='uniform', activation='relu'))
```

```python
classifier.add(Dense(output_dim=1, init='uniform',activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(X_train,y_train,batch_size=10,epochs=50, verbose=1)
```

```python
classifier.summary()
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred[:5]
```

```python
y_pred = (y_pred > 0.5) # 0.5 threshold
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```
